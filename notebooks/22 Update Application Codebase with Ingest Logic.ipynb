After exploring the logic to get files from gharchive and upload to s3, we need to update the application code base so that we can deploy and schedule it.

* Add bookmarking logic to **util/bookmark.py**.
* Add core logic to get get files from gharchive using requests and write to s3 as part of **ghactivity_ingest.py**.
* Make sure to export PYTHONPATH using `export PYTHONPATH=app` (considering you are in **ghactivity-aws** folder)
* Export required environment variables **BUCKET_NAME** and **FOLDER** and then run using `python -c "import app;app.ghactivity_ingest_to_s3()"`.
